```python
x = 0
y = 0
z = 0
df = pd.DataFrame(data)
for i in df.index:
  x += df["X"][i]
  y += df["Y"][i]
  z += df["Z"][i]

print(x,y,z)
                  
```

```python
import pandas as pd

data = pd.read_csv("Shaft_Motion_c.csv")
data
```

```python
t_data = pd.read_excel("Shaft_Translation.xlsx")
t_data
```

```python
df_t = pd.DataFrame(t_data)
```

```python
import matplotlib.pyplot as plt
import numpy as np


df_t.plot(kind='scatter',x='Count',y='X',color='red')
mymodel = np.poly1d(np.polyfit(df_t['Count'],df_t['X'],3))
myline = np.linspace(1, 2200, 10000)
plt.plot(myline,mymodel(myline))
plt.title("Variation of X with time")
plt.show()
```

```python

df_t.plot(kind='scatter',x='Count',y='Y',color='red')
plt.title("Variation of Y with time")
plt.show()
```

```python

df_t.plot(kind='scatter',x='Count',y='Z',color='red')
plt.title("Variation of Z with time")
plt.show()
```
